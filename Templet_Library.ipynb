8. Write a C++ Program to implement push and pop methods from stack using
template.

```
#include <iostream>
using namespace std;

template <class T>
class Stack
{
    private:
        T arr[5];
        int top;

    public:
        Stack()
        {
            top = -1;
        }


        void push(T value)
        {
            if (top == 4)
                {
                    cout << "Stack Overflow\n";
                }
            else
                {
                    top = top + 1;
                    arr[top] = value;
                    cout << value << " pushed onto stack\n";
                }
        }

        T pop()
        {
            if (top == -1)
                {
                    cout << "Stack Underflow\n";
                    return T();
                }
            else
                {
                    return arr[top--];
                }
        }


        bool isEmpty()
        {
            return top == -1;
        }
};

int main()
{
    Stack<int> stack;

    stack.push(10);
    stack.push(20);

    cout << "Popped value: " << stack.pop() << endl;

    if (stack.isEmpty())
        {
        cout << "Stack is empty\n";
        }
    else
        {
            cout << "Stack is not empty\n";
        }

    return 0;
}


```

9. Remove all consecutive duplicate elements from the list | C++ STL.

```

#include <iostream>
#include <list>
using namespace std;

int main()
{
    list<int> ls = {0, 1, 2, 3, 4, 2, 2, 2, 2};
    for(auto i = ls.begin(); i!= ls.end(); i++)
    {
        auto iter = i;
        iter++;

        if (iter != ls.end() && *i == *iter)
        {
            ls.erase(iter);
            --i;
        }
        else
        {}
    }

    for(auto i = ls.begin(); i!= ls.end(); i++)
    {
        cout<<*i<<" ";
    }
    return 0;
}

```

4. Reverse the Words of a String using Stack.


```
Example:
Input: str = â€œI Love To Codeâ€
Output: Code To Love I
```

```
#include <iostream>
#include <stack>
#include <string>
using namespace std;

int main()
{
    string str, result = "";
    stack<char> st;

    cout << "Enter a string: ";
    getline(cin, str);
    for (char c : str)
    {
        if (c == ' ')
        {
            while (!st.empty())
            {
                result += st.top();
                st.pop();
            }
            result += ' ';
        }
        else
        {
            st.push(c);
        }
    }

    // Process the last word
    while (!st.empty())
    {
        result += st.top();
        st.pop();
    }

    cout << "Reversed string: ";
    for (int i = result.size() - 1; i >= 0; --i)
    {
        cout << result[i];
    }
    cout << endl;

    return 0;
}


```

```
#include<iostream>
#include<stack>
using namespace std;

int main()
{
    stack<int> st;
    for (int i = 1; i < 4; i++)
    {
        st.push(i);
        st.push(i * 2);
    }

    cout << "original stack arrangement: ";
    stack<int> temp = st;
    while (!temp.empty())
    {
        cout << temp.top() << " ";
        temp.pop();
    }

    stack<int> stk;
    while (!st.empty())
    {
        auto temp = st.top();
        st.pop();

        while (!stk.empty() && stk.top() < temp)
        {
            st.push(stk.top());
            stk.pop();
        }
        stk.push(temp);
    }

    cout << endl << "sorted stack arrangement: ";
    while (!stk.empty())
    {
        cout << stk.top() << " ";
        stk.pop();
    }

    cout << endl;
    return 0;
}

```

```
Example:
Input: arr[] = [ 4 , 5 , 2 , 25 ]
Output: 4 â€“> 5
5 â€“> 25
2 â€“> 25
25 â€“> -1
```

```
#include <iostream>
#include <stack>
using namespace std;

void printNGE(int arr[], int n)
{
    stack<int> st; // Stack to store elements and find NGE

    for (int i = 0; i < n; i++)
    {
       
        while (!st.empty() && st.top() < arr[i])
        {
            cout << st.top() << " --> " << arr[i] << endl;
            st.pop(); // Pop the smaller element as we've found its NGE
        }

        st.push(arr[i]);
    }

    while (!st.empty())
    {
        cout << st.top() << " --> -1" << endl;
        st.pop();
    }
}

int main()
{
    int arr[] = {4, 5, 2, 25};
    int n = sizeof(arr) / sizeof(arr[0]);

    printNGE(arr, n);

    return 0;
}
//logic 2: sort it in assending order point nex if not equal; when equal to size() show -1;
```

```
#include<iostream>
#include<vector>
using namespace std;

int main()
{
    vector<pair<int, int>> vp = {{1, 5}, {1, 6}, {3, 4}, {2, 6}};
    vp.push_back({2, 7});
    vp.push_back({5, 8});

    for (int i = 0; i < vp.size() - 1; i++)
    {
        for (int j = i + 1; j < vp.size(); j++)
        {
            if (vp[i].first > vp[j].first)
            {
                swap(vp[i], vp[j]);
            }
        }
    }

    for (int i = 0; i < vp.size() - 1; i++)
    {
        for (int j = i + 1; j < vp.size(); j++)
        {
            if (vp[i].first == vp[j].first && vp[i].second < vp[j].second)
            {
                swap(vp[i], vp[j]);
            }
        }
    }

    for (auto x = vp.begin(); x != vp.end(); x++)
    {
        cout << "key  : " << x->first << endl;
        cout << "value: " << x->second << endl << endl;
    }

    return 0;
}


```

```
Examples:
Input: A[]={{1, 2}, {2, 2}, {3, 1}}, N = 3
Output: 2
Explanation: The longest subsequence satisfying the conditions is of length 2 and consists of {1, 2} and {3, 1};
Input: A[] = {{1, 3}, {2, 5}, {3, 2}, {5, 2}, {4, 1}}, N = 5
Output: 3
```

```
#include<iostream>
#include<deque>
using namespace std;

int main()
{
    deque<pair<int, int>> dqp = {{1, 3}, {2, 5}, {3, 2}, {5, 2}, {4, 10}};
    dqp.push_back({5, 9});
    dqp.push_back({6, 8});
    dqp.push_back({7, 5});
    dqp.push_back({8, 3});

    int result_f1 = 0;
    int result = 0;

    for(int i = 0; i < dqp.size(); i++)
    {
        if(dqp[i].first < dqp[i+1].first || dqp[i].first == dqp[i+1].first)
        {
            if(dqp[i].second > dqp[i+1].second || dqp[i].second == dqp[i+1].second)
            {
              result_f1++;
            }
        }
        else
        {
            if(result_f1 >= result)
            {
                result = result_f1;
            }
            result_f1 = 0;
        }
    }
    if(result>0)
    {
        result = result + 1;
    }
    cout<<"The longest subsequence satisfying the conditions are : "<<result;
    return 0;
}



```

5. Given an array arr[] consisting of N integers and an integer X, the task is to perform integer division on the array elements by X and print the indices of the array in non-decreasing order of their quotients obtained.
```
Examples:
Input: N = 3, X = 3, order[] = {2, 7, 4}
Output: 1  2
Explanation:
After dividing the array elements by 3, the array modifies to {0, 2, 1}. Therefore, the required order of output is 1 3 2.
Input: N = 5, X = 6, order[] = {9, 10, 4, 7, 2}
Output: 3 5 1 2 4
Explanation:
After dividing the array elements by 6, the array elements modify to 1 1 0 1 0.
Therefore, the required sequence is 3 5 1 2 4.
```

```
#include<iostream>
#include<deque>
using namespace std;

int main()
{
    int n, x, d_element;
    cout<< "Enter the size: ";
    cin>>n;

    cout<< "Enter the divisor: ";
    cin>>x;

    deque<int> dqp_original;
    deque<pair<int, int>> dqp_res;

    cout<< "Enter the elements for dequeue : ";

    for(int i = 0; i < n; i++)
    {
        cin>>d_element;
        dqp_original.push_back(d_element);
        dqp_res.push_back(make_pair(i+1, d_element/x));
    }
    cout<<"Size: "<< dqp_original.size()<<endl<<endl;
    cout<<"Original Array: "<<endl;

    for(int i = 0; i < dqp_original.size(); i++)
    {
        cout<< dqp_original[i]<< " ";
    }

    dqp_original.clear();
    dqp_original.shrink_to_fit();

    for(int i = 0; i < dqp_res.size(); i++)
    {
        for(int j = i+1; j < dqp_res.size(); j++)
        {
            if(dqp_res[i].second > dqp_res[j].second)
            {
                swap(dqp_res[i], dqp_res[j]);
            }
            else if(dqp_res[i].second == dqp_res[j].second && dqp_res[i].first > dqp_res[j].first )
            {
                swap(dqp_res[i], dqp_res[j]);
            }
        }
    }

    cout<<endl<<"After dividing and sorting array"<<endl;

    for(int i = 0; i < dqp_res.size(); i++)
    {
        cout<<dqp_res[i].first<< " "<<dqp_res[i].second<<endl;
    }
    return 0;
}


//O(n*n)
```

2. Given a string, Find the 1st repeated word in a string using unordered_map.
```
Example:
Input : "Ravi had been saying that he had been there"
Output : had
Input : "Ravi had been saying that"
Output : No Repetition
```

```
#include <iostream>
#include <unordered_map>
#include <sstream>
#include <string>
using namespace std;

string firstRepeatedWord(const string &s)
{
    unordered_map<string, int> wordCount;
    stringstream ss(s);
    string word;

    while (ss >> word)
    {
        if (wordCount[word] > 0)
        {
            return word;
        }
        wordCount[word]++;
    }
    return "No Repetition";
}

int main()
{
    string str = "Ravi had been saying that he had been there";
    cout << firstRepeatedWord(str) << endl;

    str = "Ravi had been saying that";
    cout << firstRepeatedWord(str) << endl;
}



```

5. Given a positive integer N, the task is to print the nearest power of 2 of the frequencies of each digit present in N. If there exists two nearest powers of 2 for any 3frequency, print the larger one using unordered_map.
```
Example:
Input: N = 344422
Output:
2 -> 2
3 -> 1
4 -> 4
Explanation:
Frequency of the digit 3 is 1. Nearest power of 2 is 1.
Frequency of the digit 4 is 3. Nearest power of 2 is 4.
Frequency of the digit 2 is 2. Nearest power of 2 is 2.
```

```
#include <iostream>
#include <unordered_map>
#include <cmath>

using namespace std;

int nearestPowerOfTwo(int n)
{
    if ((n & (n - 1)) == 0)
    {
        return n;
    }
    int lower = pow(2, floor(log2(n)));
    int higher = pow(2, ceil(log2(n)));
    return (n - lower <= higher - n) ? lower : higher;
}


int main()
{
    int N;
    int unit_place;
    cout << "Enter sequence: ";
    cin >> N;
    unordered_map<int, int> um;

    for (int i = N; i != 0; i = i / 10)
    {
        unit_place = N % 10;
        N = N / 10;

        auto it = um.find(unit_place);
        if (it != end(um))
        {
            it->second++;
        }
        else
        {
            um.insert({unit_place, 1});
        }
    }

    for (const auto &entry : um)
    {
        int freq = entry.second;
        int nearestPower = nearestPowerOfTwo(freq);
        cout << entry.first << " -> " << nearestPower << endl;
    }

    return 0;
}


```

Numbers from the given range whose difference is K using unordered_map.
```
Example:
Input: L = 1, R = 19, K = 6
Output: (5, 11) (7, 13) (11, 17) (13, 19)
Explanation: The pairs of prime numbers with difference 6 are (5, 11), (7, 13), (11,
17), and (13, 19).
```

```
#include <iostream>
#include <unordered_map>
#include <cmath>
using namespace std;

bool isPrime(int n)
{
    if (n <= 1) return false;
    for (int i = 2; i <= sqrt(n); ++i)
    {
        if (n % i == 0) return false;
    }
    return true;
}

void findPrimes(int L, int R, unordered_map<int, bool> &primeMap) {
    for (int i = L; i <= R; ++i)
    {
        if (isPrime(i))
        {
            primeMap[i] = true;
        }
    }
}

void printPrimePairs(int L, int R, int K)
{
    unordered_map<int, bool> primeMap;
    findPrimes(L, R, primeMap);
    for (const auto &entry : primeMap)
    {
        int prime1 = entry.first;
        int prime2 = prime1 + K;

        if (primeMap.find(prime2) != primeMap.end())
        {
            cout << "(" << prime1 << ", " << prime2 << ") ";
        }
    }
}

int main()
{
    int L, R, K;
    cout << "Enter L, R, and K: ";
    cin >> L >> R >> K;

    printPrimePairs(L, R, K);

    return 0;
}


```

10. Given an array arr[] consisting of N integers, the task is to find the maximum element
with the minimum frequency using unordered_map.
```
Example:
Input: arr[] = {2, 2, 5, 50, 1}
Output: 50
Explanation:
The element with minimum frequency is {1, 5, 50}. The maximum element among these element is 50.
```

```

#include <iostream>
#include <unordered_map>
#include <limits.h>

using namespace std;

int main()
{
    int minValue = INT_MAX;
    unordered_map<int, int> um;
    int arr[] = {2, 2, 5, 50, 1};
    int arr_length = sizeof(arr) / sizeof(arr[0]);

    for (int i = 0 ; i <arr_length ; i++)
    {
        auto it = um.find(arr[i]);
        if (it != end(um))
        {
            it->second++;
        }
        else
        {
            um.insert({arr[i], 1});
        }
    }

    for (const auto& pair : um)
    {
        if (pair.second < minValue)
        {
            minValue = pair.second;
        }
    }

    cout<<"The element with minimum frequency is/are"<<endl;
    for (const auto &entry : um)
    {
        if(entry.second == minValue)
        {

            cout<<entry.first << " ";
        }
    }

    return 0;
}

```
