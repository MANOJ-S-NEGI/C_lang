1. Define a function to input variable length string and store it in an array without memory wastage.

```

#include <stdio.h>
#include <stdlib.h>
#include <string.h>

int main()
{
    int size = 20;
    char *array;
    int current_length = 0;

    array = (char *)malloc(size * sizeof(char));
    if (array == NULL)
        {
            printf("Memory allocation failed!\n");
            return 1;
        }
        
    printf("Enter a string: ");
    fgets(array, size, stdin);
    
    current_length = strcspn(array, "\n");
    if (array[current_length] == '\n')
        {
            array[current_length] = '\0';
        }
    else
        {
            size *= 2;
            array = (char *)realloc(array, size * sizeof(char));
            if (array == NULL)
                {
                    printf("Memory reallocation failed!\n");
                    return 1;
                }
        
            if (fgets(array + current_length, size - current_length, stdin) != NULL)
                {
                    array[current_length + strcspn(array + current_length, "\n")] = '\0';
                }
        }

    printf("You entered: %s\n", array);

    free(array);
    return 0;
}

```

3. Define a class Factorial and define an instance member function to find the Factorial of a number using class.


```
#include <iostream>
using namespace std;

class Factorial
 {
    private:
        int sum;
        int calculate_factorial(int fact_num)
            {
                sum = 1;
                for (int i = 1; i <= fact_num; ++i)
                    {
                        sum *= i;
                    }
                return sum;
            }

    public:
        int display(int fact_num)
             {
                return calculate_factorial(fact_num);
            }
};

int main()
 {
    int a, fact;
    Factorial obj;

    cout << "Enter a number to print the factorial: ";
    cin >> a;

    fact = obj.display(a);
    cout << "Factorial of " << a << " is " << fact << endl;

    return 0;
}

```

```
# include <iostream>
using namespace std;

class complex
    {
        private:
            int a, b;

        public:
            complex(int x, int y)
                {
                    a = x;
                    b = y;
                }

            complex operator+(complex c)
                {
                    complex temp(0, 0);
                    temp.a = a + c.a;
                    temp.b = b + c.b;
                    if(temp.b < 0)
                        {
                            temp.b = -temp.b;
                        }
                    return temp;
                }

            complex operator-(complex c)
                 {
                    complex temp(0, 0);
                    temp.a = a - c.a;
                    temp.b = b - c.b;
                    if(temp.b < 0)
                        {
                            temp.b = -temp.b;
                        }
                    return temp;
                }

            complex operator*(complex c)
                 {
                    complex temp(0, 0);
                    temp.a = a * c.a - b * c.b;
                    temp.b = a * c.b + b * c.a;
                    if(temp.b < 0)
                        {
                            temp.b = -temp.b;
                        }
                    return temp;
                }


           bool operator==(complex c)
                {
                    return (a == c.a && b == c.b);
                }

            void display()
                {
                    cout << a << " + " << b << "i" << endl;
                }
                
};

int main()
    {
        complex c1(2, 3);
        complex c2(2, 3);

        cout << "First Complex Number: ";
        c1.display();

        cout << "Second Complex Number: ";
        c2.display();

        complex result_add = c1 + c2;
        cout << "Result of Addition: ";
        result_add.display();

        complex result_sub = c1 - c2;
        cout << "Result of Subtraction: ";
        result_sub.display();

        complex result_mul = c1 * c2;
        cout << "Result of Multiply: ";
        result_mul.display();

        if (c1 == c2)
            {
                cout << "Both numbers are the same." << endl;
            }
        else
            {
                cout << "Both numbers are not the same." << endl;
            }

        return 0;
    }

```

(Uppercase to Lowercase and vice versa).



```
#include<iostream>
#include<cstring>
using namespace std;

class mystring
    {
        private:
            char str[100];

        public:
            void input()
                {
                    cout << "Enter a string: ";
                    cin.getline(str, 100);
                }

            void display()
                {
                    cout << "String: " << str << endl;
                }

           mystring operator!()
                {
                    mystring temp;
                    strcpy(temp.str, str);

                    for(int i=0;  i<strlen(temp.str); i++)
                        {
                            if(isupper(temp.str[i]))
                                {
                                   temp.str[i] = tolower(temp.str[i]);
                                }
                            else if(islower(temp.str[i]))
                                {
                                    temp.str[i] = toupper(temp.str[i]);
                                }
                        }
                    return temp;
                }
    };

int main() {
    mystring obj, toggledObj;
    obj.input();
    obj.display();

    toggledObj = !obj;
    cout << "Toggled case: ";
    toggledObj.display();

    return 0;
}


```

```
#include<iostream>
#include<cstdlib>
using namespace std;

class palindrome
{
    private:
        int* array;
        int size;
        int iter;
        bool ispalindrome;

        int calculate_size(int input)
            {
                int count = 0;
                for(int i = input; i != 0; i /= 10)
                    {
                        count++;
                    }
                return count;
            }

    public:
        palindrome()
            {
                array = nullptr;
                size = 0;
                iter = 0;
                ispalindrome = false;
            }

        bool find_palindrome()
            {
                for(int i=0; i<iter; i++)
                    {
                        int temp_first = array[i];
                        int temp_last = array[size -1-i];
                        if(temp_first != temp_last )
                            {
                                 return false;
                            }
                         else
                            {
                               continue;
                            }
                    }
                 return true;
            }

        bool input()
            {
                int a;
                cout << "Enter a number: ";
                cin >> a;
                size = calculate_size(a);
                array = (int*)malloc(size * sizeof(int));

                if (array == nullptr)
                    {
                        cout << "Memory allocation failed!" << endl;
                        return 0;
                    }

                cout << "size is: " << size << endl;

                for(int i = size - 1; i >= 0; i--)
                    {
                        array[i] = a % 10;
                        a = a / 10;
                    }

                if (size % 2 == 0 )
                    {
                        iter = size/2;
                    }
                else
                    {
                        iter = (size/2) +1;
                    }
                 cout << "iter size: " << iter << endl;

                 bool ispalindrome  = find_palindrome();
             }

        void display()
            {
                ispalindrome?cout << "Number Entered is palindrome ":cout << " Entered Number is not a palindrome ";
            }

        ~palindrome()
            {
                delete[] array;  // Free dynamically allocated memory
            }
};

int main()
{
    palindrome obj;
    obj.input();
    obj.display();
    return 0;
}

```
